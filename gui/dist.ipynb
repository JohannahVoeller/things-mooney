```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

metadata = pd.read_csv('/Users/johannahvoeller/Documents/Projects/things-mooney/things_mooney_metadata.csv')
```

```python
# Plot the pre and post subjective recognition histograms
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(15, 5))
fig.suptitle('Metadata Distribution for THINGS-Mooney')

# Subjective Recognition
sns.histplot(metadata.mean_subj_recog_pre, kde=True, stat="density", label="Pre", ax=ax1, color=sns.color_palette("Paired")[4])
sns.histplot(metadata.mean_subj_recog_post, kde=True, stat="density", label="Post", ax=ax1, color=sns.color_palette("Paired")[5])
ax1.set_title("Mean subjective recognition rate")
ax1.set_xlabel("Subjective recognition (%)")

# Semantic Distance
sns.histplot(metadata.mean_sem_dist_pre, kde=True, stat="density", label="Pre", ax=ax2, color=sns.color_palette("Paired")[6])
sns.histplot(metadata.mean_sem_dist_post, kde=True, stat="density", label="Post", ax=ax2, color=sns.color_palette("Paired")[7])
ax2.set_title("Mean semantic distance")
ax2.set_xlabel("Semantic Distance")

# Semantic Entropy
sns.histplot(metadata.pre_semantic_entropy, kde=True, stat="density", label="Pre", ax=ax3, color=sns.color_palette("Paired")[8])
sns.histplot(metadata.post_semantic_entropy, kde=True, stat="density", label="Post", ax=ax3, color=sns.color_palette("Paired")[9])
ax3.set_title("Mean semantic entropy")
ax3.set_xlabel("Semantic entropy")

# Add legend
fig.legend(labels=["Pre", "Post"], title="Recognition Stage")
```

```python
# Define an aesthetically pleasing color scheme for the variables
colors = {
    "subjective_recognition": {"pre": "#FFD1DC", "post": "#FF69B4"},  # Pink shades
    "semantic_distance": {"pre": "#C1E1C1", "post": "#228B22"},  # Green shades
    "semantic_entropy": {"pre": "#ADD8E6", "post": "#1E90FF"},  # Blue shades
}

# Plot the pre and post distributions
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Metadata Distribution for THINGS-Mooney', fontsize=16)

# Subjective Recognition
sns.histplot(metadata.mean_subj_recog_pre, kde=True, stat="density", 
             ax=ax1, color=colors["subjective_recognition"]["pre"])
sns.histplot(metadata.mean_subj_recog_post, kde=True, stat="density", 
             ax=ax1, color=colors["subjective_recognition"]["post"])
ax1.set_title("Mean Subjective Recognition Rate", fontsize=12)
ax1.set_xlabel("Subjective Recognition (%)", fontsize=10)
ax1.tick_params(axis='both', which='major', labelsize=9)

# Semantic Distance
sns.histplot(metadata.mean_sem_dist_pre, kde=True, stat="density", 
             ax=ax2, color=colors["semantic_distance"]["pre"])
sns.histplot(metadata.mean_sem_dist_post, kde=True, stat="density", 
             ax=ax2, color=colors["semantic_distance"]["post"])
ax2.set_title("Mean Semantic Distance", fontsize=12)
ax2.set_xlabel("Semantic Distance", fontsize=10)
ax2.tick_params(axis='both', which='major', labelsize=9)

# Semantic Entropy
sns.histplot(metadata.pre_semantic_entropy, kde=True, stat="density", 
             ax=ax3, color=colors["semantic_entropy"]["pre"])
sns.histplot(metadata.post_semantic_entropy, kde=True, stat="density", 
             ax=ax3, color=colors["semantic_entropy"]["post"])
ax3.set_title("Mean Semantic Entropy", fontsize=12)
ax3.set_xlabel("Semantic Entropy", fontsize=10)
ax3.tick_params(axis='both', which='major', labelsize=9)

# Add a single legend for all subplots
""" fig.legend(
    labels=[
        "Pre (Subjective Recognition)", "Post (Subjective Recognition)",
        "Pre (Semantic Distance)", "Post (Semantic Distance)",
        "Pre (Semantic Entropy)", "Post (Semantic Entropy)"
    ],
    title="Legend",
    title_fontsize=12,
    loc="upper center", ncol=2, fontsize=10,
    bbox_to_anchor=(0.5, 1.05)
) """

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Leave space for the legend and title
plt.show()
```

```python
# Plot the pre and post subjective recognition histograms
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(15, 5))
fig.suptitle('Bivariate Distributions for Metadata Variables Pre- vs. Post-Disambiguation')

# Subjective Recognition
sns.kdeplot(metadata, x="mean_subj_recog_pre", y="mean_subj_recog_post", ax=ax1)
ax1.set_title("Subjective Recognition")
ax1.set_xlabel("Pre-Disambiguation")
ax1.set_ylabel("Post-Disambiguation")

# Semantic Distance
sns.kdeplot(metadata, x="mean_sem_dist_pre", y="mean_sem_dist_post", ax=ax2)
ax2.set_title("Semantic Distance")
ax2.set_xlabel("Pre-Disambiguation")
ax2.set_ylabel("Post-Disambiguation")

# Semantic Entropy
sns.kdeplot(metadata, x="pre_semantic_entropy", y="post_semantic_entropy", ax=ax3)
ax3.set_title("Semantic Entropy")
ax3.set_xlabel("Pre-Disambiguation")
ax3.set_ylabel("Post-Disambiguation")
```
