# Visualization of Data Distribution for [THINGS-Mooney](https://github.com/wobc/things-mooney)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

metadata = pd.read_csv('things_mooney_metadata.csv')
```

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Plot the pre and post subjective recognition histograms
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Metadata Distribution for THINGS-Mooney', fontsize=16, fontweight='bold')

# Subjective Recognition
sns.histplot(metadata.mean_subj_recog_pre, kde=True, stat="density", label="Pre", ax=ax1, color=sns.color_palette("Paired")[4], linewidth=2, linestyle='-', alpha=0.7)
sns.histplot(metadata.mean_subj_recog_post, kde=True, stat="density", label="Post", ax=ax1, color=sns.color_palette("Paired")[5], linewidth=2, linestyle='--', alpha=0.7)
ax1.set_title("Mean Subjective Recognition Rate", fontsize=14)
ax1.set_xlabel("Subjective Recognition (%)", fontsize=12)
ax1.set_ylabel("Density", fontsize=12)
ax1.legend(labels=["Pre", "Post"], title="Recognition Stage")

# Semantic Distance
sns.histplot(metadata.mean_sem_dist_pre, kde=True, stat="density", label="Pre", ax=ax2, color=sns.color_palette("Paired")[6], linewidth=2, linestyle='-', alpha=0.7)
sns.histplot(metadata.mean_sem_dist_post, kde=True, stat="density", label="Post", ax=ax2, color=sns.color_palette("Paired")[7], linewidth=2, linestyle='--', alpha=0.7)
ax2.set_title("Mean Semantic Distance", fontsize=14)
ax2.set_xlabel("Semantic Distance", fontsize=12)
ax2.set_ylabel("Density", fontsize=12)
ax2.legend(labels=["Pre", "Post"], title="Recognition Stage")

# Semantic Entropy
sns.histplot(metadata.pre_semantic_entropy, kde=True, stat="density", label="Pre", ax=ax3, color=sns.color_palette("Paired")[8], linewidth=2, linestyle='-', alpha=0.7)
sns.histplot(metadata.post_semantic_entropy, kde=True, stat="density", label="Post", ax=ax3, color=sns.color_palette("Paired")[9], linewidth=2, linestyle='--', alpha=0.7)
ax3.set_title("Mean Semantic Entropy", fontsize=14)
ax3.set_xlabel("Semantic Entropy", fontsize=12)
ax3.set_ylabel("Density", fontsize=12)
ax3.legend(labels=["Pre", "Post"], title="Recognition Stage")

# Tight layout to avoid overlaps
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Plot the pre and post subjective recognition histograms
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(18, 6))
fig.suptitle('Bivariate Distribution for Metadata Variables Pre- vs. Post-Disambiguation', fontsize=16, fontweight='bold')

# Subjective Recognition
sns.kdeplot(metadata, x="mean_subj_recog_pre", y="mean_subj_recog_post", ax=ax1, color=sns.color_palette("Paired")[5])
ax1.set_title("Subjective Recognition", fontweight='bold')
ax1.set_xlabel("Pre-Disambiguation")
ax1.set_ylabel("Post-Disambiguation")

# Semantic Distance
sns.kdeplot(metadata, x="mean_sem_dist_pre", y="mean_sem_dist_post", ax=ax2, color=sns.color_palette("Paired")[7])
ax2.set_title("Semantic Distance", fontweight='bold')
ax2.set_xlabel("Pre-Disambiguation")
ax2.set_ylabel("Post-Disambiguation")

# Semantic Entropy
sns.kdeplot(metadata, x="pre_semantic_entropy", y="post_semantic_entropy", ax=ax3, color=sns.color_palette("Paired")[9])
ax3.set_title("Semantic Entropy", fontweight='bold')
ax3.set_xlabel("Pre-Disambiguation")
ax3.set_ylabel("Post-Disambiguation")

# Tight layout to avoid overlaps
plt.tight_layout()

# Show plot
plt.show()
```
